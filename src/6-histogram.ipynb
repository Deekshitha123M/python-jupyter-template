### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with a 2x2 layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Data for line chart
x = np.linspace(0, 10, 100)
y = x ** 2
axes[0, 0].plot(x, y, label='y = x^2', color='b')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values')
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple'])
axes[1, 1].set_title('Pie Chart')

# Data for histogram
fig, ax = plt.subplots(figsize=(6, 4))
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
ax.hist(data, bins=5, color='teal', edgecolor='black')
ax.set_title('Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
